# Use Taylor to approximate functions at a given point
given:
$$f(1) = 2; f'(1) -3; f''(1) = 4; f^{(n)}(1)=0 \ \ \forall n\geq3$$
evaluate f(4)
$$f(4) = \sum_{i=0}^{\infty}f^{(i)}\frac{(x_0)}{i!}(x-x_0) \\
= f(x_0)+f'(x_0)(x-x_0)+f''(x_0)/2*(x-x_0)^2 +0 \\
= f(1)+f'(1)(4-1)+f''(1)/2*(4-1)^2 \\
= 2+-3(3)+4/2*(3^2)\\
= 11$$

given: $$f(x) = \frac{1}{(20x-10)}$$
write the taylor approximation of degree=2 about point $x_0 = 0$
$$
f'(x) = \frac{-1(20)}{(20x-10)^2} -> f'(0) = -1/5 \\
f''(x) = \frac{20*2*(20x-10)*20}{(20x-10)^4} -> f''(0) = -4/5 \\
t_2(x) = -f(0)+x*f'(0)-x^2/2*f''(0) \\= -1/10-1/5x-x^2/2*-4/5
$$
