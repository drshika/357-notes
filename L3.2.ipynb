**Significant Digits**

0.0000350 has 3 sig figs   

$\hat{x}$ has **n** significant figures of $x$ if $| \hat{x} - x|$ has zeroes in the first n decimal places counting from the leftmost non-zero (leading) digit of $x$ followed by a digit from *0 to 4*

# Big O Notation

$$f(x) = O(g(x)) \text{ as } x\rightarrow a$$
Iff there exists a value $M$ and $\delta$ such that
$$|f(x) \leq M|g(x)| \text{ } \forall x \text{ where } x < |s-a| < \delta$$

Suppose the computational complexity of a numerical method is given by $O(n^3)$. When n = 1000, it was observed that the method takes 10 seconds to complete.
You would like to run the same method for n = 10000. What is an estimate of the time for completion of the larger problem?

$$t = cn^3 \\
n_1 = 1000 -> t_1 = 10s \\
t_1 = cn_1^3 \\
t_2 = cn_2^3 \\
\frac{t_1}{t_2} = \frac{n_1^3}{n_2^3} \\
t_2 = (\frac{n_2}{n_1})^3t_1
$$

```python
n_1 = 1000
n_2 = 10000
(n_2/n_1)**3*10
```

```python
# solve for c
t_1 = 1024
n_1 = 8
c = t_1/(n_1**3)
c
```

```python
# use c to get n
t_2 = 9826
n_2 = (t_2 / c)**(1/3)
n_2
```
