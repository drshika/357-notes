# Fixed Point
 **Range:** difference between the largest and smallest numbers possible.
 More bits for the integer part == increased range   
 **Precision:** smallest possible difference between any two numbers.
 More bits for the fractional part == greater precision.    
 
 We can just let the binary point float

# Floating Point Numbers
Represents a range of numbers with the same number of fixed digits.

float can be expressed as $x = \pm q * 2^m = \pm b_0.b_1...b_n * 2^m$

where q is the significand (normally a fractional value inthe range \[1.0, 2.0)   
m is the exponent: $m \in [L, U]$ from [-4, 4] for example

$b_i \in {0,1}$   
**Precision:** p = n+1

# Normalized Floating Point
$$x = \pm 1.b_1b_2...b_n*2^m = \pm 1.f*2^m$$
where $f$ is the franctional part of the exponent and $b_i\in{0,1}$
- in the original representation of a 5 bit number, we had 5 bits of precision
- in this representation, we store the 1 in a different location but can use the 5 bits of storage to store an additional bit so we gain one bit of precision (6 bits of precision total). 
- **Exponent Range:** $m \in [L, U]$
- **Precision:** $p = n+1$
- **Smallest positive normalized FP number:** $2^L$
- **Largest positive normalized FP number:** $2^{u+1}(1-2^{-p})$


```python
# Smallest Numbner
L = -4
2**L
```

```python
# Largest Number
u = 4
n = 2
p = n + 1
2**(u+1)*(1-2**(-p))
```

**Machine Epsilon:** $\epsilon_m$ distance between 1 and the next largest floating point number. 
$$\epsilon_m = 2^{-n}$$
**Largest number you can represent:** $2^p$

```python
u = 14
n = 4
p = n + 1
2**(u+1)*(1-2**(-p))
```
